## Function to generate all Institute data 

```python
def generate_all_data(num_rows):
    data={
        "Institute Name": [],
        "Phone Number" : [],
        "Email Address":[],
        "Location": [],
        "Owner Name":[],
        "No. of Courses":[]
       
    }
    for i in range(num_rows):
        institute_name = random.choice(institute_names)
        phone_number = f"+91-98765412{i%100:02d}"
        email_address = f"contact{(i+1)%100}@institute{i%50}.com"
        location =random.choice(locations)
        owner_name=random.choice(owner_names)
        no_of_courses = random.randint(i,50)
        
        data["Institute Name"].append(institute_name)
        data["Phone Number"].append(phone_number)
        data["Email Address"].append( email_address)
        data["Location"].append( location)
        data["Owner Name"].append(owner_name)
        data["No. of Courses"].append(no_of_courses)
        
    return pd.DataFrame(data)
```

## Generate 50 rows of all Institute  data

```python
df =generate_all_data(50)
```

## Draw Scatter Plot

```python
plt.figure(figsize=(8,6))
for location in df['Location'].unique():
    subset = df[df['Location'] ==location]
    plt.scatter(subset['Location'],subset['No. of Courses'],label=location)
plt.title('Number of Courses Available by Location')
plt.xlabel("Location")
plt.ylabel('No.of Couses Available')
plt.legend()
plt.show()
plt.grid()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df ,x='Location')
plt.title('No.of Institute per Location')
plt.xlabel("Location")
plt.ylabel("No.of Institute")
plt.show()
```

## Draw Bar Plot

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df ,x='Location')
plt.title('No.of Institute per Location')
plt.xlabel("Location")
plt.ylabel("No.of Institute")
plt.xticks(rotation=45)
plt.show()
```
